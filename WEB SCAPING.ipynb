importer les packages

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

saisir l'URL

```python
url="https://feeds.bbci.co.uk/news/rss.xml"
```

importer le code de la page

```python
reponse= requests.get(url)
soup= BeautifulSoup(reponse.text,"html.parser")
```

```python
items=soup.findAll('item')
```

```python
items
```

```python
news_item=[]
```

```python
for i in items:
    news_i={}
    news_i['title']=i.title.text
    news_i['description']=i.description.text
    news_i['pubdate']=i.pubdate.text
    news_item.append(news_i)
```

```python
news_i
```

convertir le dictionnaire en base de données

```python
df=pd.DataFrame(news_item, columns=['title','description','pubdate'])
```

```python
df.head()
```

```python
df.shape
```

convertir en csv

```python
df.to_csv('web_scapping.csv',index=False, encoding='utf-8')
```
